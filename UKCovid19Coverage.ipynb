```python

import matplotlib
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.ticker as ticker

from matplotlib.dates import DateFormatter
from matplotlib.pyplot import figure

filename = '/content/sample_data/UKCovid19Data.csv'

# Setting up dataframe
df = pd.read_csv(filename)
age_range = df['Age Range'].unique()

# Setting up chart area
fig, axes = plt.subplots(nrows = 6, ncols = 1, figsize=(15,50));

for n in range (6):
  
  # Creating filtered dataframes
  df_age_range = df.loc[df['Age Range'] == age_range[n], :]

  # Start creating line graph
  axes[n].plot(df_age_range['Covid Vacc Date'], df_age_range['BlackCum']/df_age_range['Black total'], label='Black')
  axes[n].plot(df_age_range['Covid Vacc Date'], df_age_range['MixedCum']/df_age_range['Mixed total'], label='Mixed')
  axes[n].plot(df_age_range['Covid Vacc Date'], df_age_range['OtherCum']/df_age_range['Other total'], label='Other')
  axes[n].plot(df_age_range['Covid Vacc Date'], df_age_range['SouthAsianCum']/df_age_range['South Asian total'], label='South Asian')
  axes[n].plot(df_age_range['Covid Vacc Date'], df_age_range['UnknownCum']/df_age_range['Unknown total'], label='Unknown')
  axes[n].plot(df_age_range['Covid Vacc Date'], df_age_range['WhiteCum']/df_age_range['White total'], label='White')

  # Formatting tickers and labels
  axes[n].yaxis.set_major_locator(plt.MaxNLocator(10))
  axes[n].xaxis.set_major_locator(plt.MaxNLocator(20))
  axes[n].yaxis.set_major_formatter(ticker.PercentFormatter(1))
  axes[n].legend(loc='upper left', )
  axes[n].locator_params(axis='y', nbins=10)
  axes[n].title.set_text('UK Covid 19 Vaccination Status - ' + age_range[n] + ' years old')
  axes[n].tick_params(labelrotation=45)

plt.savefig('UK_Covid19_Vaccination_Status.png', dpi = 300)
```
